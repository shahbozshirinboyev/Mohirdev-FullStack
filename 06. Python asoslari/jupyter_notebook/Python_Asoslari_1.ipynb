# Ro'yxatlar (Lists)

Biz shu paytgacha o'zgaruvchilarga bitta qiymat yuklashni o'rgandik. Biz hozir yangi ma'lumot turi ya'ni List (Ro'yxat) bilan tanishamiz:

```python
# Bitta o'zgaruvchiga bir nechta qiymatlarni (ya'ni ro'yxatni) yuklash
mevalar = ['Olma', 'Anjir', 'Shaftoli', "O'rik"]
narxlar = [12000, 18000, 34000, 8000]

print(mevalar)
print(narxlar)

# Ro'yxat ichida har qanday turdagi ma'lumotlarni saqlash imkoniyati bor.
royxat = [-23, "olma", "talaba", 45]
print(royxat)
```

```python
# Bo'sh bo'lgan ro'yxat ham yaratish imkoniyati bor
ismlar = []
print(ismlar)
```

Ro'yxat ichidagi har bir qiymat element deyiladi. Har bir elementga biz tartib raqami bo'yicha murojaat qilishimiz mumkin.

```python
mevalar = ['Olma', 'Anjir', 'Shaftoli', "O'rik"]
# Ro'yxat elementlarini indexlashda 0 qiymatdan boshlanadi doim
print(mevalar[0]) # Birinchi element
print(mevalar[3]) # Oxirgi element
```

```python
# Ro'yxatlardagi elementlarni oxiridan boshlab ham chaqirishimiz mumkin:
# Oxirgi element -1, undan oldingisi -2, hakozo ....
mevalar = ['Olma', 'Anjir', 'Shaftoli', "O'rik"]

print(mevalar[-1]) #oxirgi element
print(mevalar[-4]) #birinchi element

# Agar juda ham uzun ro'yxat bo'ladigan bo'lsa uning oxirgi elementini chaqirib olish uchun bu juda yaxshi usul bo'ladi.
```

```python
mevalar = ['Olma', 'Anjir', 'Shaftoli', "O'rik"]
# Mevalar ro'yxati elementlari string bo'lganligi uchun uning elementlariga stringga aloqador methodlarni qo'llash mumkin.
print(mevalar[0].upper())
print(mevalar[0].lower())
```

```python
narxlar = [12000, 18000, 34000, 8000]
print(narxlar[0] + narxlar[1])
print(narxlar[0] + narxlar[1] - 50000)
```

# Ro'yxatlar bilan iishlash. O'zgarmas ro'yxatlar (Tuples)

Qanday qilib ro'yxat elementlarini alifbo tartibida tartiblash.
.sort() - metodi

```python
cars = ['bmw', 'audi', 'volvo', 'general motors', 'tesla']
cars.sort()
print(cars)
```

Pythonda katta va kichik harflar farqlanganligi sababli agar ro'yxat ichidagi elementlar katta harflar yoki kichik harflar bilan boshlanadigan bo'lsa birinchi bo'lib katta harflar sort() qilinadi keyin esa kichik harflar.

```python
cars = ['Bmw', 'audi', 'volvo', 'general motors', 'tesla']
cars.sort()
print(cars)
```

```python
# Agar jadvalni a harfidan emas alifboning oxirgi harfidan boshlab bartiblash kerak bo'lsa unda
# .sort(reverse=True) dan foydalanamiz z->a ko'rinishida bo'ladi.
cars = ['bmw', 'audi', 'volvo', 'general motors', 'tesla']
cars.sort(reverse=True)
print(cars)
```

```python
# Asl ro'yxatga hech qanday o'zgartirish kiritmasdan sorted(royxat) funksiyasi orqali ro'yxatni tartiblash imkoni bor.
cars = ['bmw', 'audi', 'volvo', 'general motors', 'tesla']
print(sorted(cars))
print(sorted(cars, reverse=True))
```

```python
# Sonli ro'yxatlar uchun ham .sort() methodi va sorted() funksiyasidan foydalanishimiz mumkin.
sonlar = [12, 34, 432, 1, -5, -7.3]
print(sorted(sonlar))
print(sorted(sonlar, reverse=True))
```

Ro'yxat ichidagi maximum qiymatni topish uchun max() funksiyasidan foydalaniladi.

```python
toq_sonlar = list(range(1, 31, 2))
print(toq_sonlar)
max_qiymat = max(toq_sonlar)
print("Max: ", max_qiymat)

# min() bilan esa minimum qiymatni topish imkoniyati bor:
min_qiymat = min(toq_sonlar)
print("Min: ", min_qiymat)

# ro'yxat ichidagi elementlarni qiymatlarnini yig'indisini topish uchun esa
# sum() - funksiyasidan foydalaniladi
yigindi = sum(toq_sonlar)
print("Summa: ", yigindi)
```

Ro'yxat elementlarini ma'lum bir qismdan boshqa bir qismgacha ajratib olish uchun (Kesib olish uchun) royxat[first_index:last_index] ko'rinishida foydalaniladi.

```python
cars = ["bmw", "audi", "volvo", "general motors", "tesla"]
my_wish_cars = cars[0:2]
print(my_wish_cars)

# yoki
print(cars[0:3])

# Bunda ham e'tibor beradigan narsa cars[0:3] deganda 3-indexdagi element yangi ro'yxatga kirmaydi.
# Birinchi qiymatni bermasak o'rniga nol oladi va ro'yxat boshidan olishni boshlaydi
# cars[:3] bu cars[0:3] bilan teng!
print(cars[:3])

# agar oxirigi indexni ko'rsatmasak ro'yxatni birinchi ko'rsatilgan indexidan
# oxirgi indexigacha qirqib oladi
print(cars[2:])
```

```python
# ro'yxatdan nusxa olish
cars = ["bmw", "tesla", "volvo"]
my_cars = cars # E'tibor beraylik bu nusxa olish emas, shunchaki bitta ro'yxatga ikki xil nom berish
print(my_cars)
cars.append("audi")
print(my_cars)
# ro'yxatdan nusxa olish quyidagicha:
my_cars = cars[:]
cars.remove('audi')
print(cars)
print(my_cars)

# Demak ko'rib turganimizdek nusxa olish kesish usulidan foydalanishdek bo'ladi.
```

Ro'yxatning yangi turi TUPLE

```python
toys = ('bus', 'car', 'bear', 'dino', 'snake', 'lizard')
print(toys[0])
print(toys[-1])
print(toys[:3])

# Lekin bu ro'yxat ichiga yangi element qo'shib yoki element qiymatini o'zgartirib bo'lmaydi
# Chunki bu o'zgarmas ro'yxat hisoblanadi
```

```python
# Agar ro'yxatni o'zgartirishga yoki boshqa element qo'shish yoki o'chirish kerak bo'lsa
# uni avval oddiy ro'yxatga almashtirib olish kerak bo'ladi
toys = ('bus', 'car', 'bear', 'dino', 'snake', 'lizard')
print(toys)
toys = list(toys)
print(toys)
toys.append("teddy")
print(toys)
# va oxirida yana o'zgarmas turdagi ro'yxat kerak bo'lsa tuple()
# funksiyasidan foydalanib uni o'z turiga qaytaish mumkin
toys = tuple(toys)
print(toys)
```

```python
countries = ['Brazil', 'America', 'Canada', 'India']
print(countries)
print("Ro'yxat uzunligi: ", len(countries))
print("Tartiblangan ro'yxat: ", sorted(countries))
print("Tartiblangan ro'yxat: ", sorted(countries, reverse=True))
print(countries)
countries.reverse()
print(countries)
countries.sort()
print(countries)
countries.sort(reverse=True)
print(countries)
```
